## Импорты

```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.sparse import csr_matrix, identity, tril, triu, diags, SparseEfficiencyWarning
from scipy.sparse.linalg import inv, eigs

from typing import Tuple

import warnings
warnings.simplefilter('ignore', SparseEfficiencyWarning)
```

# Реализация

Сначала реализуем все необходимые функции для решения задания, а потом построим графики. 

## Построение матрицы `A`

Будем хранить матрицу в сжатом виде. 

```python
def build_sparse_matrix_A(N: int) -> csr_matrix:
    h = 1.0 / (N + 1)
    points = [i * h for i in range(1, N + 1)]
    
    data, row, col = [], [], []
    
    done = np.zeros(N ** 2)
    
    val = -1.0 / h ** 2
    
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            k = i + (j - 1) * N
            
            if done[k - 1] == 0:
                done[k - 1] = 1
                data.append(4.0 / h ** 2 + np.exp(points[i - 1] * points[j-1]))
                row.append(k-1)
                col.append(k-1)
            
            if i > 1:
                data.append(val)
                row.append(k-1)
                col.append(i + (j - 1) * N - 2)
            if i < N:
                data.append(val)
                row.append(k-1)
                col.append(i + (j - 1) * N)

            if j > 1:
                data.append(val)
                row.append(k-1)
                col.append(i + (j - 2) * N - 1)
            if j < N:
                data.append(val)
                row.append(k-1)
                col.append(i + j * N - 1)

                
    return csr_matrix((data, (row, col)), (N ** 2, N ** 2))
```

## Поиск максимального собственного числа

```python
def get_max_lambda(N: int, A: csr_matrix = None, iters: int = 10000, eps: float = 1e-16) -> float:
    if A is None:
        A = build_sparse_matrix_A(N)
    
    u_0 = np.ones((N ** 2, 1))
    
    lambda_cur, lambda_prev = None, None
    
    for k in range(iters):
        u_0 = u_0 / np.linalg.norm(u_0)
        
        u_1 = A @ u_0
        t = (u_1.T @ u_0) / (u_0.T @ u_0)
        
        lambda_cur, lambda_prev = t[0][0], lambda_cur
        
        u_0 = u_1
        
        if k > 1 and np.abs(lambda_cur - lambda_prev) < eps:
            break

    return lambda_cur
```

## Поиск минимального собственного числа

```python
def get_min_lambda(N: int, A: csr_matrix = None, iters: int = 1000) -> Tuple:
    if A is None:
        A = build_sparse_matrix_A(N)
    
    max_lambda = get_max_lambda(N, A) 
    alpha = max_lambda + 100
    D = A - alpha * identity(N ** 2)
    
    min_lambda = get_max_lambda(N, D) + alpha
    
    return min_lambda, max_lambda / min_lambda
```

## Метод Зейделя

```python
def seidel_method(N: int, A: csr_matrix, iters: int = 10000) -> float:
    L = tril(A, k=0)
    U = triu(A, k=1)
    Z = inv(L) @ U

    return np.abs(get_max_lambda(N, Z, iters))
```

## Метод Якоби

```python
def jacobi_method(N: int, A: csr_matrix, iters: int = 10000) -> float:
    D = diags(A.diagonal())
    B = A - D
    D_inv = inv(D)
    Z = D_inv @ B
    
    return np.abs(get_max_lambda(N, Z, iters))
```

# Графики

Так как во всех заданиях требуется изучить зависимость от параметра $N$, а в лекции выше - это размер матрицы, то на всех графиках по оси абцисс откладываем значения $N^2$, то есть количество строк / столбцов матрицы `A`. А само значение $N$ - это параметр размера сетки, которая используется для построения самой матрицы (как в прошлом задании).

Посмотрим на зависимость максимального собственного числа от значения $N^2$.

```python
%%time
N_grid = np.arange(10, 111, 10)

eigenvalues = []
error = []
    
for N in N_grid:
    A = build_sparse_matrix_A(N)
    eigenvalue = get_max_lambda(N, A, iters=100000)
    eigenvalues.append(eigenvalue)
    error.append(np.abs(eigs(A, k=1, return_eigenvectors=False) - eigenvalue))
```

```python
plt.figure(figsize=(16,8))
    
x = [i ** 2 for i in N_grid]
plt.plot (x, eigenvalues, linewidth=3)

plt.xlabel('$N^2$, количество строк матрицы A', fontsize=15)
            
plt.ylabel('Значение макс. собственного числа', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.grid()
title = 'Зависимость значения макс. с. ч. от $N^2$ при нахождении методом прямых итераций'
plt.title(title, fontsize=15)
plt.show()
```

Как видно на графике выше, с ростом размера матрицы, увеличивается и значение максимального собственного числа. Причем эта зависимость максимального собственного числа от $N^2$ - количества строк - линейна.

Посмотрим на то, как точно мы находим значение максимального собственного числа. Для этого сравним полученное значение с истинным, найденным с помощью `scipy`.

```python
plt.figure(figsize=(16,8))
    
x = [i ** 2 for i in N_grid]
plt.plot (x, error, color='r', linewidth=3)

plt.xlabel('$N^2$, количество строк матрицы A', fontsize=15)
            
plt.ylabel('|Error|', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.grid()
title = 'Погрешность вычисления макс. собственного числа'
plt.title(title, fontsize=15)
plt.show()
```

Находим мы весьма точно, как видно, с ростом размера матрицы требуется все больше итераций, для достижения лучшей точности.

## Поиск минимального собственного числа

## Методы Зейделя, Якоби

```python
%%time
N_grid = np.arange(3, 51, 1)

lst_seidel = []
lst_jacobi = []
for N in N_grid:
    A = build_sparse_matrix_A(N)
    lst_seidel.append(1 - seidel_method(N, A))
    lst_jacobi.append(1 - jacobi_method(N, A))
```

```python
x = [i ** 2 for i in N_grid]

plt.figure(figsize=(16,8))
plt.plot (x, lst_seidel, linewidth=3)

plt.xlabel('$N^2$, количество строк матрицы A', fontsize=15)
plt.ylabel('1 - q', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.grid()
plt.title('Метод Зейделя', fontsize=15)
plt.show()
```

```python
x = [i ** 2 for i in N_grid]

plt.figure(figsize=(16,8))
plt.plot (x, lst_jacobi, linewidth=3)

plt.xlabel('$N^2$, количество строк матрицы A', fontsize=15)
plt.ylabel('1 - q', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.grid()
plt.title('Метод Якоби', fontsize=15)
plt.show()
```

```python
x = [i ** 2 for i in N_grid]

plt.figure(figsize=(16,8))
plt.plot (x, lst_seidel, label='Метод Зейделя', linewidth=3)
plt.plot (x, lst_jacobi, label='Метод Якоби', linewidth=3)

plt.xlabel('$N^2$, количество строк матрицы A', fontsize=15)
plt.ylabel('1 - q', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.title('1 - q для методов Зейделя и Якоби', fontsize=15)
plt.show()
```

Посмотрим в лог-формате.

```python
x = [i ** 2 for i in N_grid]

plt.figure(figsize=(16,8))
plt.plot (x, np.log10(lst_seidel), label='Метод Зейделя', linewidth=3)
plt.plot (x, np.log10(lst_jacobi), label='Метод Якоби', linewidth=3)

plt.xlabel('$N^2$, количество строк матрицы A', fontsize=15)
plt.ylabel('$\log(1 - q)$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.title('$\log(1 - q)$ для методов Зейделя и Якоби', fontsize=15)
plt.show()
```

```python
x = [i ** 2 for i in N_grid]

plt.figure(figsize=(16,8))
plt.plot (np.log10(x), np.log10(lst_seidel), label='Метод Зейделя', linewidth=3)
plt.plot (np.log10(x), np.log10(lst_jacobi), label='Метод Якоби', linewidth=3)

plt.xlabel('$\log(N^2)$', fontsize=15)
plt.ylabel('$\log(1 - q)$', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)
plt.grid()
plt.title('$\log(1 - q)$ для методов Зейделя и Якоби', fontsize=15)
plt.show()
```
